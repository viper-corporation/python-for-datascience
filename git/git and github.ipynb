## git and github

### what is git?

Git is a **free version control system tool**. And it is a tool you can download and install on your machine, no matter if it's a Windows or macOS or Linux machine and with that Git tool, **you can manage source code changes of your projects**. **any source code can be managed with Git**, no matter which programming language is used or which kind of application is being built. 

for this purpose you can do multiple things with git :  
 
1. allowing you to **create snapshots**, so-called commits of that source code at any point of time you want to create a snapshot.
2. You can also work on multiple versions of the same package or app or website or whatever simultaneously with help of a feature called **branches.**
3. And you can easily switch between different branches and snapshots by **checking them out**

> **you can simply download git and install it**

### what is git repository?

A Git repository is the .git/ folder inside a project. This repository **tracks all changes** made to files in your project, building a history over time. Meaning, **if you delete the .git/ folder, then you delete your projectâ€™s history.**

### working with git commands

As mentioned before already, Git is mainly about those code **snapshots** or **commits**. And the idea is that as a developer, once you made some meaningful changes to your code, once you finish the work on a certain feature or on a bug fix or anything like this, you create a new snapshot, (you create a new commit) And doing that is actually a two-step process:

1. first by running ``git add`` and pointing at the files that contain the changes that should be included
in that snapshot because you control which files and which changes should be part of a given snapshot.
This process is called **staging**. You stage the changes that should be included in the next commit.
```bash
git add <file(s)>
```

2. after running git add,you have to run ``git commit.`` This then creates a commit based on the changes
that have been staged.
```bash
git commit -m "your message of this commit"
```

over time, you can create more and more commits
to reflect your different code changes, new features and various bug fixes you might have implemented.

And the idea behind having all these different commits and this history of commits is of course that you can then go back to an older commit whenever you need to, for example, to revert a change.

And you can see the current status of your Git repository and for example, find out which files are staged and which files are not staged
by running ``git status``, another key command. This command doesn't change anything but it gives you some information about the current status of this Git repository.

```bash
git status
```

You can  run ``git log`` to get a list, a chronologically ordered list of all the commits you have thus far and at the moment,

```bash
git log
```

after you run git log you can see your commits and **commit id** and the git pointer which called ``HEAD``.And this ``HEAD`` pointer simply tells Git which state is currently loaded

You can move between those commits with the ``git checkout`` command, which needs the ID of the commit you wanna move to,and that command would then allow you to temporarily move to another commit,which means the other commits that come after the commit you're moving to
won't be deleted. You just temporarily load another state.

```bash
git checkout <command id>
```

If I wanna get back to my branch that includes all the commits, including the commit which I'm now not seeing anymore here, I can simply type 

```bash
git checkout main.
```

### who to delete commits?

one option would be to manually make the changes we wanna make, like, for example, removing that paragraph
and then adding a new commit. That is something you can do and that is absolutely fine.
But it's actually more work than you need to do because Git also offers various commands that help with undoing commits. For example, the very important git revert command. This also needs the ID of a commit you wanna revert and what this command does then
is it takes a look at the commit with that ID that you're specifying here and it then reverts the changes that were made by that commit. So not all the changes since that commit but the changes that are defined in that commit with that ID.

> it reverted by adding a new commit.

```bash
git revert <commit id>
```

But sometimes you really want to delete a commit, you wanna get rid of a commit, and you can do that with the git reset --hard command.
This also needs the ID of a commit but here it's the ID of a commit you wanna go back to.
And all the commits since that commit you wanna go back to will be deleted. That's what git reset --hard does.

```bash
git reset --hard <commit id>
```

### how to ignore tracking files by git?

Git supports a special file,which you can add to your project, a ``.gitignore file``.
**this file must be named .gitignore.**
This file allows you to specify folder and file names of folders and files that should always be ignored when you run git add.


### Branches

as you're working on a project,
you start collecting commits. You commit your various changes and eventually, you have more and more commits.
And these commits are organized in branches, and you got one branch out of the box without doing anything. **The main branch, it could also be called master branch though.**
But you do have this default branch because that branch is created whenever you initialize a new Git repository
and you start creating commits.

you can have more than one branch in a project.You can create additional branches, for example to work on a new feature or new bug

You could create a new feature branch
by ``git branch`` command. By default, this command takes the latest command, which is currently targeted by head and it builds a new branch with this commit

eventually you might want to combine all these code changes though
because for example if we want create a website, the goal is, not to build multiple websites
but to still work on one single website just on different features or bug fixes
that are all relevant to that same website. And that's why eventually,
you can merge those different branches with the git ``merge command.``
If you run that command, all the commits that are part of one branch of the branch,  will be merged with the commits of the main branch. And Git will automatically try to commit those changes

commands you should now for branches and merge:

```bash
git branch <branch name> 
git branch
git checkout <branch name>
git branch -D <branch name>
git checkout -b <branch name>
git merge <branch name>
```

## github

So we used Git locally on our machine, and that's the idea behind Git.
It is a tool you use locally on a machine. We created commits, worked with branches,
and now we have this history of commits here. Now, of course, if you're only working locally
with Git, that has a couple of disadvantages. For example, if your computer crashes
and you can't recover it, your data, and this includes your development projects,
will be lost. Collaborating with team members or colleagues is also unnecessarily hard
because you basically need to manually copy the code across different machines. For example, with USB sticks or as attachments on emails
or anything like that. And the same is true if you ever need to switch machines.
If you're traveling and you have a different laptop with you than you normally use for development,
you don't have immediate access to your code. That's why we have GitHub.

> remember : **Git is a independent tool and GitHub is a independent company.**

 the key feature offered by GitHub is its repositories feature.
You can create those cloud storage buckets in your GitHub account,
and then connect your local Git repositories to those GitHub repositories
to back up your code or use it across different devices and so on.m

### how to connect git repository to github repository?

Once you create it, a GitHub repository,
no matter if it's public or private, you connect it so to say
to your local Git repository by running the ``git remote`` command, to be precise, the git remote add command typically.
This adds a remote repository, like this GitHub repository, to your local repository.
It establishes a connection. You can then thereafter start ``pushing`` your local commits
to that cloud repository so that they are stored there. And you can download commits, which you don't have on your machine yet with ``git pull.``

```bash
git remote add origin <repository url>
git push origin master
git pull
```

### git config

before you push your code you need to set git config. ``Git config`` is a powerful Git command that allows you to customize how Git works and optimize it to fit your workflow.

 As shown below, Git pulls configuration settings from a narrowing list of files. Each configuration level in the graphic below can overwrite the configuration settings in the level above it. For instance, Git config global settings overwrite Git config system settings, but not Git config local settings. 

Now that you have a general understanding of the different Git config levels, letâ€™s take a look at how you can use Git config commands to set your username and email. Git requires that a username and email be established before you can leverage all of its capabilities, so itâ€™s a good idea to get this out of the way immediately after you download Git. 

To use Git config username to set your username, navigate to a terminal and run: 
```bash
git config --global user.name "Your Name"
```

To use Git config email to set your email in the terminal, run: 
```bash
git config --global user.email youremail@example.com
```
